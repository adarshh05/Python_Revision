Lambda Functions in Python

Lambda functions are small anonymous functions defined using the **lambda** keyword.they can have any number of arguments but only one expression. They are commonly used for short
operations or as arguments to higher-order functions.

```python
def addition(a,b):
    return a + b

print(addition(5,4))
```

```python
addition = lambda a, b: a + b

print(addition(5, 4))
```

```python
def triple_number_addition(x, y, z):
    return x + y + z

print(triple_number_addition(5, 5, 5))
```

```python
triple = lambda x, y, z: x + y + z

print(triple(5, 5, 5))
```

```python
#map function can be used to put the function to more values like in a list

numbers = [2, 4, 6, 7, 9]
def square(numbers):
    return numbers **2

square(2)
```

```python
list(map(lambda x: x **2, numbers))
```
