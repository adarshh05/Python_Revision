```python
def even_or_odd(num):
    '''this function finds even or odd'''
    if num%2==0:
        print("Then number is even")
    else:
        print("The number is odd")
```

```python
## function with multiple parameters

def add(a,b):
    return a + b

result = add(2,4)
print(result)
```

```python
def greet(name):
    print(f"Hello {name} welcome to the paradise")

greet('krish')
```

```python

# So this is the default parameter in which if in any case the name is not written then it goes to the paranthesis and
# sees that whether the defualt parameter is given or not and if it is given then instead of the name it takes the default parameter
# name name puts on the output


def greet(name= 'Guest'):
    print(f"Hello {name} welcome to the paradise")

greet()
```

```python
# variable length argument
# positonal and keyword arguments

def print_numbers(*args):
    for numbers in args:
        print (numbers)
```

```python
print_numbers(1, 2, 3, 4, 5, 6, 7, "Adarsh")
```

```python
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(name="Adarsh", age= "25", city= "Mumbai")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f"positional arguments= {val}")
    
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(1, 2, 3, 4, 5, "adarsh", name= "adarsh", age= "20", country= "India")
```
