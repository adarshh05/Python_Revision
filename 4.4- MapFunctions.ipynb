**The Map() Function in Python**

The map() function applies a given function to all items in an input list(or any other iterable) and returns a map
object(an iterator). This is particularly useful for transforming data in a list comprehensively

```python
def squares(numbers):
    return numbers**2

print(squares(5))
```

```python
numbers= [1, 3, 5, 7, 9]
list(map(squares, numbers))
```

```python
# using lamda with map function

numbers= [1, 3, 5, 7, 9]
list(map(lambda X: X*X, numbers))
```

```python
numbers1= [1, 2, 3]
numbers2= [4, 5, 6]

added_numbers= list(map(lambda x, y: x+y, numbers1, numbers2))

print(added_numbers)
```

```python
words = ['apple', 'banana', 'cherry']
upper_word = list(map(str.upper, words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people = [{'name': 'Adarsh', 'age': 20 },
          {'name': 'John', 'age': 30}]

only_names=list(map(get_name, people))
print(only_names)
```
